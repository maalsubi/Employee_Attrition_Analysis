CREATING A DASHBOARD OF THE ABOVE PLOTS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import panel as pn

# Set up the plotting environment
sns.set(style="whitegrid")
pn.extension('plotly')

# 1. Stacked Area Chart
def stacked_area_chart():
    df_grouped = data.groupby(['YearsAtCompany', 'Attrition']).size().unstack()
    ax = df_grouped.plot(kind='area', stacked=True, figsize=(3, 4), alpha=0.7)
    ax.set_title('Stacked Area Chart of Years at Company vs Attrition')
    ax.set_ylabel('Count of Employees')
    ax.set_xlabel('Years at Company')
    return ax.get_figure()

# 2. Multiple Line Chart
def multiple_line_chart():
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.lineplot(x='YearsAtCompany', y='MonthlyIncome', hue='Attrition', data=data, ax=ax)
    ax.set_title('Monthly Income over Years at Company by Attrition')
    ax.set_ylabel('Monthly Income')
    ax.set_xlabel('Years at Company')
    return fig

# 3. Scatterplot
def scatterplot():
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.scatterplot(x='Age', y='MonthlyIncome', hue='Attrition', data=data, ax=ax)
    ax.set_title('Scatterplot of Age vs Monthly Income by Attrition')
    ax.set_ylabel('Monthly Income')
    ax.set_xlabel('Age')
    return fig

# 4. Distplot/Histplot
def distplot():
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.histplot(data=data, x='MonthlyIncome', hue='Attrition', kde=True, element="step", ax=ax)
    ax.set_title('Histogram and KDE of Monthly Income by Attrition')
    ax.set_ylabel('Density')
    ax.set_xlabel('Monthly Income')
    return fig

# 5. Donut Chart
def donut_chart():
    attrition_counts = data['Attrition'].value_counts()
    fig, ax = plt.subplots(figsize=(3, 4))
    wedges, texts, autotexts = ax.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
    for text in texts:
        text.set_color('grey')
    for autotext in autotexts:
        autotext.set_color('white')
    ax.set_title('Donut Chart of Attrition')
    return fig

# 7. Descriptive Statistics
def descriptive_statistics():
    return data.describe().to_html()

# 8. Countplot
def countplot():
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.countplot(x='Department', hue='Attrition', data=data, ax=ax)
    ax.set_title('Countplot of Department by Attrition')
    ax.set_ylabel('Count')
    ax.set_xlabel('Department')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    return fig

# 9. Histogram
def histogram():
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.histplot(data=data, x='Age', hue='Attrition', kde=True, element="step", ax=ax)
    ax.set_title('Histogram and KDE of Age by Attrition')
    ax.set_ylabel('Density')
    ax.set_xlabel('Age')
    return fig

# 10. Scatter Plot
def scatter_plot():
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.scatterplot(x='MonthlyIncome', y='Age', hue='Attrition', data=data, ax=ax)
    ax.set_title('Scatter Plot of Monthly Income vs Age by Attrition')
    ax.set_xlabel('Monthly Income')
    ax.set_ylabel('Age')
    return fig

# 11. Boxplot
def boxplot():
    fig, ax = plt.subplots(figsize=(5, 6))
    sns.boxplot(x='JobRole', y='MonthlyIncome', hue='Attrition', data=data, ax=ax)
    ax.set_title('Boxplot of Monthly Income by Job Role and Attrition')
    ax.set_ylabel('Monthly Income')
    ax.set_xlabel('Job Role')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    return fig

# 12. Violin Plot
def violin_plot():
    fig, ax = plt.subplots(figsize=(3,5))
    sns.violinplot(x='EducationField', y='MonthlyIncome', hue='Attrition', data=data, split=True, ax=ax)
    ax.set_title('Violin Plot of Monthly Income by Education Field and Attrition')
    ax.set_ylabel('Monthly Income')
    ax.set_xlabel('Education Field')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    return fig

# 13. Heatmap of Attrition Counts by Job Role and Department
def heatmap_attrition_counts():
    pivot_table = data.pivot_table(index='JobRole', columns='Department', values='Attrition', aggfunc='count').fillna(0)
    fig, ax = plt.subplots(figsize=(3, 4))
    sns.heatmap(pivot_table, annot=True, fmt='g', cmap='coolwarm', ax=ax)
    ax.set_title('Heatmap of Attrition Counts by Job Role and Department')
    ax.set_xlabel('Department')
    ax.set_ylabel('Job Role')
    return fig

# Create Panel layout
# Create CSS styles
css_styles = """
<style>
/* General Container Styling */
.grid-container {
    display: grid;
    grid-template-columns: repeat(3, 1fr);
    gap: 20px;
    padding: 20px;
    background-color: #f5f5f5;
    border-radius: 10px;
    box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);
}

/* Grid Item Styling */
.grid-item {
    background-color: #ffffff;
    padding: 15px;
    border-radius: 8px;
    box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);
    transition: transform 0.2s ease, box-shadow 0.2s ease;
    position: relative;
}

.grid-item:hover {
    transform: scale(1.03);
    box-shadow: 0px 6px 12px rgba(0, 0, 0, 0.2);
}

.grid-item img {
    border: 1px solid #ddd;
    border-radius: 5px;
}

/* Header Styling */
.grid-header {
    grid-column: span 3;
    font-size: 28px;
    font-weight: bold;
    color: #333333;
    padding-bottom: 10px;
    border-bottom: 2px solid #cccccc;
    margin-bottom: 20px;
}

/* Responsive Design */
@media (max-width: 1200px) {
    .grid-container {
        grid-template-columns: repeat(2, 1fr);
    }
}

@media (max-width: 800px) {
    .grid-container {
        grid-template-columns: 1fr;
    }
}

/* Custom Styling for Markdown Headers */
.grid-header h1 {
    font-size: 36px;
}

.grid-header h2 {
    font-size: 30px;
}

/* Add Padding to HTML Descriptive Statistics */
.grid-item p {
    margin: 0;
    font-size: 14px;
    line-height: 1.6;
}
</style>
"""
# Create the Panel dashboard with grid layout and custom styles
dashboard = pn.Column(
    pn.pane.HTML(css_styles),
    pn.pane.HTML('<div class="grid-container">', sizing_mode='stretch_width'),
    pn.pane.Markdown("# Employee Attrition Analysis", css_classes=['grid-header']),
    
    pn.pane.Matplotlib(stacked_area_chart(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(multiple_line_chart(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(scatterplot(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(distplot(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(donut_chart(), tight=True, css_classes=['grid-item']),
    
    pn.pane.HTML(descriptive_statistics(), css_classes=['grid-item']),
    pn.pane.Matplotlib(countplot(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(histogram(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(scatter_plot(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(boxplot(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(violin_plot(), tight=True, css_classes=['grid-item']),
    pn.pane.Matplotlib(heatmap_attrition_counts(), tight=True, css_classes=['grid-item']),
    
    pn.pane.HTML('</div>', sizing_mode='stretch_width')
)
# Serve the dashboard
dashboard.servable()

```

TRAINING AN ML MODEL TO PREDICT 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Convert categorical variables to numerical
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    if column != 'Attrition':  # Skip encoding the target variable for now
        label_encoders[column] = LabelEncoder()
        data[column] = label_encoders[column].fit_transform(data[column])

# Encode target variable 'Attrition'
data['Attrition'] = data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Split data into features and target
X = data.drop(columns=['Attrition', 'EmployeeID'])  # Exclude 'EmployeeID' and target variable
y = data['Attrition']

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for better readability
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)

# Display the shapes of the resulting datasets
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

# Feature importance
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance)
plt.title('Feature Importance')
plt.show()
```
